```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def gen_stock_price_arr(tdays=100):
    price_arr = np.ones(tdays)
    for ind in np.arange(0, tdays-1):
        if ind == 0:
            win = np.random.binomial(1, 0.5)
        else:
            win = price_arr[ind] > price_arr[ind - 1]
        price_arr[ind+1] = price_arr[ind] * (1.05 if win else 0.95)
    return price_arr
```

```python
# plt.plot(gen_stock_price_arr())
_, axes = plt.subplots(nrows=1, ncols=2, figsize=(14, 4))
axes[0].plot(gen_stock_price_arr())
axes[1].plot(gen_stock_price_arr())
plt.show()
```

```python
# 连续上涨3天后的下跌概率为55%
# 连续下跌3天后的上涨概率为80%

pre_price_arr = gen_stock_price_arr()
# 以之前模拟的价格位置为起始点
def gen_stock_price_arr(tdays=251):
    price_arr = np.concatenate(pre_price_arr, np.ones(tdays) * pre_price_arr[-1])
    for ind in np.arange(0, tdays-1):
        if ind == 0:
            win = np.random.binomial(1, 0.5)
        else:
            win = price_arr[ind] > price_arr[ind - 1]
        price_arr[ind+1] = price_arr[ind] * (1.05 if win else 0.95)
    return price_arr
```
