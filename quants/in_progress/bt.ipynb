```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# np.arange(12000)
def mm(group):
    return (group - group.min())/(group.max() - group.min())
```

```python
g = np.arange(12000)
mm(g)
_, splots = plt.subplots(nrows=3, ncols=1, figsize=(20,8))
splots[0].plot(mm(g))
splots[1].plot(2*np.sqrt(mm(g))-1)
splots[2].plot(np.power(mm(g), 4))
plt.show()
```

```python
g = np.arange(10000)
plt.plot(np.power(mm(g), 4))
plt.show()
```
